### **STRING**

```python
# Example string
my_string = "   Hello, World!   "

# 1. len(): Get the length of a string.
length = len(my_string)
print("1. Length:", length)  # Output: 17
print(f"1. Length: {length}")

# 2. Case-related methods
lower_case = my_string.lower()
upper_case = my_string.upper()
capitalized = my_string.capitalize()
title_case = my_string.title()
print("2. Lowercase:", lower_case)      # Output: "   hello, world!   "
print("   Uppercase:", upper_case)      # Output: "   HELLO, WORLD!   "
print("   Capitalized:", capitalized)    # Output: "   hello, world!   "
print("   Title Case:", title_case)      # Output: "   Hello, World!   "

# 3. Strip methods
stripped = my_string.strip()
left_stripped = my_string.lstrip()
right_stripped = my_string.rstrip()
print("3. Stripped:", stripped)          # Output: "Hello, World!"
print("   Left Stripped:", left_stripped) # Output: "Hello, World!   "
print("   Right Stripped:", right_stripped) # Output: "   Hello, World!"

# 4. replace(): Replace a substring with another.
replaced = my_string.replace("World", "Universe")
print("4. Replaced:", replaced)          # Output: "   Hello, Universe!   "

# 5. Search and Find
search_result = my_string.find("World")
print("5. Find 'World' Index:", search_result)  # Output: 6

# 6. Split and Join
words = my_string.split()  # Splits by spaces (default)
joined = "-".join(words)   # Joins with a hyphen
print("6. Split Words:", words)        # Output: ['Hello,', 'World!']
print("   Joined with Hyphen:", joined)  # Output: "Hello,-World!"

# 7. Checking content methods
starts_with_hello = my_string.startswith("Hello")
ends_with_world = my_string.endswith("World!")
is_alpha_numeric = my_string.isalnum()
is_numeric = my_string.isnumeric()
print("7. Starts with 'Hello':", starts_with_hello)  # Output: False
print("   Ends with 'World!':", ends_with_world)     # Output: False
print("   Is Alpha-Numeric:", is_alpha_numeric)      # Output: False
print("   Is Numeric:", is_numeric)                  # Output: False
print("orignal  =",my_string)
```

```python
# Creating a list of integers
numbers = [1, 2, 3, 4, 5]

# Creating a list of strings
fruits = ['apple', 'banana', 'cherry']

# Checking the number of elements in the list using len()
num_elements = len(numbers)                                # Calculating the number of elements in 'numbers'

   # Adding an element to the end of the list using append()
numbers.append(6)                                          # Appending the value 6 to the end of the 'numbers' list
   # Inserting an element at a specific index using insert()
fruits.insert(1, 'orange')                                 # Inserting 'orange' at index 1 in the 'fruits' list
   # Extending a list with another list using extend()
fruits.extend(['pear', 'grape'])                           # Adding 'pear' and 'grape' to the end of the 'fruits' list

   # Removing the first occurrence of an element using remove()
numbers.remove(3)                                          # Removing the first occurrence of the value 3 from 'numbers'
   # Removing and returning an element by index using pop()
removed_fruit = fruits.pop(2)                              # Removing the element at index 2 from 'fruits' and storing it in 'removed_fruit'

# Using list comprehension to create a new list
squared_numbers = [x ** 2 for x in numbers]                # Creating a new list of squared numbers from 'numbers'

# Checking if an element is present in a list using in
is_apple_present = 'apple' in fruits                      # Checking if 'apple' is in the 'fruits' list
is_8_present = 8 in numbers  # Checking if the value 8 is in the 'numbers' list


# Reversing the order of elements in the list using reverse()
numbers.reverse()                                                                # Reversing the order of elements in the 'numbers' list

# Sorting the list in ascending order using sort()
numbers.sort()                                                                   # Sorting the elements of the 'numbers' list in ascending order

# Creating a shallow copy of a list using copy()
fruits_copy = fruits.copy()                                                      # Creating a shallow copy of the 'fruits' list

# Clearing all elements from a list using clear()
fruits.clear()                                                                   # Removing all elements from the 'fruits' list
#ONE CAN USE del[:] WHICH USED IN ALL SEQUENCES

# Printing the modified lists and results
print("Modified numbers:", numbers)  # Printing the modified 'numbers' list
print("Modified fruits:", fruits)  # Printing the modified 'fruits' list
print("Removed fruit:", removed_fruit)  # Printing the removed fruit
print("Number of elements in numbers:", num_elements)  # Printing the number of elements in 'numbers'
print("Squared numbers:", squared_numbers)  # Printing the squared numbers
print("'apple' in fruits:", is_apple_present)  # Printing the result of the membership check
print("8 in numbers:", is_8_present)  # Printing the result of the membership check
print("Fruits copy:", fruits_copy)  # Printing the copied 'fruits' list
```

```python
# Tuple Basics:
# A tuple is an ordered, immutable(you can add) collection of elements enclosed in parentheses ().

#1 Creating a Tuple:
my_tuple = (1, 2, 3, 'apple')

#2 Counting Elements:
# Use the count() method to count occurrences of a specific element.
my_tuple = (1, 2, 2, 3, 3, 3)
count_3 = my_tuple.count(3)
print(count_3)  # Output: 3 (3 appears three times)

#3 Finding Indices:
# Use the index() method to find the index (position) of an element.
my_tuple = ('apple', 'banana', 'cherry', 'apple')
index_cherry = my_tuple.index('cherry')
print(index_cherry)  # Output: 2 (index of 'cherry' is 2)

#4 Packing and Unpacking:
# Tuples can be packed and unpacked easily.

# Packing:
point = 3, 4     # no bracket = tuple
print(point)  # Output: (3, 4)

# Unpacking:
x, y = point
print(x)  # Output: 3
print(y)  # Output: 4

#5 Slicing Tuples:
# Use slicing to access subsets of a tuple.
my_tuple = (1, 2, 3, 4, 5)
subset = my_tuple[1:4]
print(subset)  # Output: (2, 3, 4)


#6 Concatenating Tuples:
# Concatenate tuples using the + operator.
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = tuple1 + tuple2
print(result)  # Output: (1, 2, 3, 4, 5, 6)

#7 Iterating Through a Tuple:
# Use a for loop to iterate through tuple elements.
my_tuple = (1, 2, 3, 4)
for item in my_tuple:
    print(item)
# Output:
# 1
# 2
# 3
# 4

# Nesting Tuples:
# Tuples can be nested within other tuples.
nested_tuple = ((1, 2), (3, 4), (5, 6))

# Tuple Unpacking with * (Extended Unpacking):
# Use * to capture multiple elements during tuple unpacking.
my_tuple = (1, 2, 3, 4, 5)
first, *middle, last = my_tuple
print(first)    # Output: 1
print(middle)   # Output: [2, 3, 4]
print(last)     # Output: 5

# Immutable Elements:
# While a tuple is immutable, its elements can be mutable (e.g., lists).
mutable_tuple = ([1, 2], [3, 4])
mutable_tuple[0].append(5)
print(mutable_tuple)  # Output: ([1, 2, 5], [3, 4])

# Named Tuples (collections.namedtuple):
# Named tuples provide named fields for tuples.
from collections import namedtuple
Person = namedtuple('Person', ['name', 'age'])
person = Person('Alice', 30)
print(person.name, person.age)  # Output: Alice 30
```

```python
# Dictionary Basics:
# A dictionary is an unordered, mutable collection of key-value pairs.
# key in dictionary is immutable
# Creating a Dictionary:
my_dict = {'name': 'Alice', 'age': 30}

# Accessing Values:
name = my_dict['name']
print("Accessing Values:")
print(name)  # Output: 'Alice'

# Modifying a Dictionary:
my_dict['city'] = 'New York'
my_dict['age'] = 31
del my_dict['age']
print("\nModifying a Dictionary:")
print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York'}

# Dictionary Methods:

# 1. get(key, default=None):
age = my_dict.get('age', 0)
print("\nUsing get(key, default=None):")
print(age)  # Output: 0 (default value since 'age' is not in the dictionary)

# 2. keys():
keys = my_dict.keys()
print("\nUsing keys():")
print(keys)  # Output: dict_keys(['name', 'city'])

# 3. values():
values = my_dict.values()
print("\nUsing values():")
print(values)  # Output: dict_values(['Alice', 'New York'])

# 4. items():
items = my_dict.items()
print("\nUsing items():")
print(items)  # Output: dict_items([('name', 'Alice'), ('city', 'New York')])

# 5. pop(key, default=None):
name = my_dict.pop('name', 'Unknown')
print("\nUsing pop(key, default=None):")
print(name)  # Output: 'Alice' (removed key 'name')

# 6. update(other_dict):
another_dict = {'country': 'USA', 'age': 32}
my_dict.update(another_dict)
print("\nUsing update(other_dict):")
print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York', 'country': 'USA', 'age': 32}

# 7. clear():
my_dict.clear()
print("\nUsing clear():")
print(my_dict)  # Output: {} (empty dictionary)

# Additional Methods:

# 8. copy():
copy_dict = my_dict.copy()
print("\nUsing copy():")
print(copy_dict)  # Output: {}

# 9. setdefault(key, default=None):
age = my_dict.setdefault('age', 25)
city = my_dict.setdefault('city', 'New York')
print("\nUsing setdefault(key, default=None):")
print(age)  # Output: 25
print(city)  # Output: 'New York'

# 10. popitem():
item = my_dict.popitem()
print("\nUsing popitem():")
print(item)  # Output: ('city', 'New York') (an arbitrary key-value pair)

# 11. fromkeys(seq, value=None):
keys = ['name', 'age', 'city']
default_value = 'Unknown'
new_dict = dict.fromkeys(keys, default_value)
print("\nUsing fromkeys(seq, value=None):")
print(new_dict)  # Output: {'name': 'Unknown', 'age': 'Unknown', 'city': 'Unknown'}

# 12. defaultdict(default_factory):
from collections import defaultdict
fruit_counts = defaultdict(int)
fruit_counts['apple'] += 1
print("\nUsing defaultdict(default_factory):")
print(fruit_counts['apple'])  # Output: 1

# 13. Dictionary Comprehension:
numbers = [1, 2, 3, 4]
squared_dict = {num: num**2 for num in numbers}
print("\nUsing Dictionary Comprehension:")
print(squared_dict)  # Output: {1: 1, 2: 4, 3: 9, 4: 16}
```

### **Set and Boolean**

```python
fruits = {"apple", "banana", "cherry"}

# Adding Elements:
fruits.add('orange')
print("Set after adding elements:", fruits)

# Removing Elements:
fruits.remove('banana')  # Raises an error if 'banana' is not found
fruits.discard('strawberry')  # Doesn't raise an error if 'strawberry' is not found
print("Set after removing elements:", fruits)

# Set Methods:

# 1. add(element)
fruits.add('kiwi')

# 2. remove(element)
fruits.remove('apple')

# 3. discard(element)
fruits.discard('strawberry')

# 4. clear()
fruits.clear()
print("Set after clear():", fruits)

# 5. copy()
fruits_copy = fruits.copy()
print("Copy of the set:", fruits_copy)

# 6. union(other_set) or |
fruits = {'apple', 'banana'}
more_fruits = {'cherry', 'orange'}
combined_fruits = fruits.union(more_fruits)
print("Combined set (union):", combined_fruits)

# 7. intersection(other_set) or &
fruits = {'apple', 'banana', 'cherry'}
common_fruits = fruits.intersection({'banana', 'kiwi'})
print("Common elements (intersection):", common_fruits)

# 8. difference(other_set) or -
fruits = {'apple', 'banana', 'cherry'}
unique_fruits = fruits.difference({'banana', 'kiwi'})
print("Unique elements (difference):", unique_fruits)

# 9. issubset(other_set)
fruits = {'apple', 'banana'}
subset = fruits.issubset({'apple', 'banana', 'kiwi'})
print("Is subset : " , subset)

# 10. issuperset(other_set)
superset = fruits.issuperset({'apple'})
print("Is superset:", superset)

# 11. pop()
popped_element = fruits.pop()
print("Popped element:", popped_element)

# 12. len()
num_fruits = len(fruits)
print("Number of elements:", num_fruits)

# 13. difference_update(other_set)
fruits = {'apple', 'banana', 'cherry'}
fruits.difference_update({'banana', 'kiwi'})
print("Set after difference_update:", fruits)

# 14. intersection_update(other_set)
fruits = {'apple', 'banana', 'cherry'}
fruits.intersection_update({'banana', 'kiwi'})
print("Set after intersection_update:", fruits)

# 15. symmetric_difference(other_set) or ^
fruits = {'apple', 'banana', 'cherry'}
new_fruits = fruits.symmetric_difference({'banana', 'kiwi'})
print("Symmetric difference:", new_fruits)

# 16. symmetric_difference_update(other_set)
fruits = {'apple', 'banana', 'cherry'}
fruits.symmetric_difference_update({'banana', 'kiwi'})
print("Set after symmetric_difference_update:", fruits)

# 17. isdisjoint(other_set)
fruits = {'apple', 'banana'}
disjoint = fruits.isdisjoint({'cherry', 'kiwi'})
print("Is disjoint:", disjoint)
```

```python
# Define a function that takes another function as an argument
def apply_operation(operation, x, y):
    result = operation(x, y)
    return result

# Define a couple of functions to be used as operations
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

# Use the apply_operation function with different operations
result1 = apply_operation(add, 5, 3)
result2 = apply_operation(subtract, 5, 3)

print(result1)  # Output: 8
print(result2)  # Output: 2
```

 **DECORATIVE**

```python
# DECORATIVE FUNCTION

def decorate(func):                           # Decorating gift
  def wrapper():
    print("before")
    func()
    print("after")
  return wrapper

@decorate
def say_hello():                               #Gift
  print(2+2,7**8,3425+324)

say_hello()
```

```python
def log_function_call(func):
    def wrapper(*a):
        print(f"Calling {func.__name__} with arguments {a} and keyword arguments ")
        return func(*a)
    return wrapper

@log_function_call      #use @
def add(a,b):
    return a + b

add(50, 60)
```

```python
import time

# Decorator function to measure execution time
def measure_time(func):
    def wrapper(*args):
        start_time = time.time()
        result = func(*args)
        end_time = time.time()
        print(f"{func.__name__} took {end_time - start_time:6f} seconds to execute.")
        # return result
    return wrapper


# Applying the decorator
@measure_time
def some_slow_function():
  print(2+2,7**8,3425+324)

    # time.sleep(3)  # Simulate a time-consuming task

@measure_time
def another_slow_function():
    time.sleep(2)

# Calling the decorated functions
some_slow_function()
another_slow_function()
```

```python
# Colab benchmark: compare 3 (plus enumerate) implementations
# Paste this whole block into one Colab cell and run.

import timeit
import pandas as pd
import matplotlib.pyplot as plt
import math

# --- Implementations to test (mirror your original code) ---
def method_index_inplace_copy(lst):
    """Index loop that modifies a copy (so original isn't mutated for benchmark fairness)."""
    l = lst.copy()
    for idx in range(len(l)):
        if l[idx] % 2 == 0:
            l[idx] = '**'
        else:
            l[idx] = '*'
    return l

def method_manual_counter(lst):
    """Manual counter + append (creates new list)."""
    a = -1
    modified_list = []
    for i in lst:
        a += 1  # unnecessary but included to mirror original
        if i % 2 == 0:
            modified_list.append("**")
        else:
            modified_list.append("*")
    return modified_list

def method_list_comprehension(lst):
    """List comprehension (creates new list)."""
    return ["**" if i % 2 == 0 else "*" for i in lst]

def method_enumerate_inplace_copy(lst):
    """Enumerate-based in-place modification on a copy (clearer than manual index)."""
    l = lst.copy()
    for idx, val in enumerate(l):
        l[idx] = '**' if val % 2 == 0 else '*'
    return l

# --- Choose sizes and benchmark parameters ---
# Default sizes are reasonable for Colab. Increase if you want slower, clearer separation.
sizes = [10_000, 50_000, 200_000]
# If you'd like to test 1_000_000, add it here but expect more runtime and memory.
# sizes = [10_000, 50_000, 200_000, 1_000_000]

repeat = 5   # number of times timeit.repeat will run (take best)
number = 3   # number of function calls per timing measurement

methods = [
    ("index_inplace_copy", method_index_inplace_copy),
    ("manual_counter_append", method_manual_counter),
    ("list_comprehension", method_list_comprehension),
    ("enumerate_inplace_copy", method_enumerate_inplace_copy),
]

rows = []
print("Starting benchmark. This may take a minute depending on sizes...\n")

for n in sizes:
    base_list = list(range(1, n + 1))
    print(f"Benchmarking size: {n}")
    for name, func in methods:
        # Use timeit.Timer with a lambda to call the function with the prepared list
        timer = timeit.Timer(lambda f=func, bl=base_list: f(bl))
        times = timer.repeat(repeat=repeat, number=number)  # list of totals for `number` runs
        best_total = min(times)  # best (fastest) total time among repeats
        per_call = best_total / number
        rows.append({
            "size": n,
            "method": name,
            "best_total_seconds_for_number_runs": best_total,
            "per_call_seconds": per_call,
            "repeat_times": times,
        })
        print(f"  {name:25} -> best per-call: {per_call:.6f}s (best total {best_total:.6f}s)")

print("\nAll benchmarks complete.\n")

# --- Create DataFrame and pivot summary ---
df = pd.DataFrame(rows)
pivot = df.pivot(index="size", columns="method", values="per_call_seconds").reset_index()
pivot.columns.name = None

# Display the numeric summary
print("Per-call best-run times (seconds):")
display(pivot)

# --- Plot results ---
plt.figure(figsize=(9, 5))
for method_name in pivot.columns[1:]:
    plt.plot(pivot['size'], pivot[method_name], marker='o', label=method_name)
plt.xscale('log')   # input sizes often plotted on log-scale for clarity
plt.yscale('log')   # runtime often behaves linearly but constants make log useful
plt.xlabel('input size (n) [log scale]')
plt.ylabel('time per call (s) [log scale]')
plt.title('Benchmark: methods runtime vs input size')
plt.legend()
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.show()

# --- Helpful derived info (optional) ---
print("\nRelative speed (per-call) at each size (normalized to list_comprehension = 1.0):")
for n in sizes:
    row = pivot[pivot['size'] == n].iloc[0]
    base = row['list_comprehension']
    ratios = {col: (row[col] / base) for col in pivot.columns if col != 'size'}
    print(f" size={n}: " + ", ".join([f"{k}={v:.2f}x" for k,v in ratios.items()]))

# --- End ---
print("\nNotes:\n- list_comprehension is typically the fastest in CPython due to C-level optimizations.\n- All methods are O(n); differences are per-item overheads and memory behavior.\n- If you want to test larger sizes, add them to the `sizes` list above.")
```

```python
def is_prime(num):
  """Checks if a number is prime."""
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

# Find prime numbers up to a certain limit
limit = 50
prime_numbers = [num for num in range(2, limit + 1) if is_prime(num)]
print(f"Prime numbers up to {limit}: {prime_numbers}")
```
